```python
import numpy as np
import xarray as xr
from Data import location, terciles, probs
da = location('Kenya_Hindcast')
obs = location('Kenya_Obs_Chirps')
```

**Change the "below" to "above","below" or "average" to veiw the alternative graphs**

```python
a,b,c,d = probs(da,obs,"average") # Enter Here


import matplotlib.pyplot as plt
b = b*100
a = a*100



#data
forecast_prob = [0.15, 0.70]  
observed_freq = [a, b]     

# Plotting the data points
plt.scatter(forecast_prob, observed_freq, color='green', label='Data points')

# Plotting the perfect reliability line
plt.plot([0, 1], [0, 100], 'k--', label='Perfect reliability')

# Adding labels and title
plt.xlabel('Forecast probability (centre of ranges: 0 to .3; .4 to 1.0)')
plt.ylabel('Observed frequency')
plt.title('Reliability diagram: ECMWF_OND_St:Aug / CHIRPS_Obs_OND / (1993-2016)')

# Adding legend
plt.legend()
plt.grid(True)
plt.xticks([0.15,  0.70])
# Displaying the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
forecast_prob = [0.15, 0.70]  
usage_counts = [d, c]        

# Plotting the bar chart
plt.figure(figsize=(8, 6))
plt.bar(forecast_prob, usage_counts, width=0.1, color='skyblue')

# Adding labels and title
plt.xlabel('Forecast probability (centre of range)')
plt.ylabel('No. of times used')
plt.title('Frequency of use diagram: ECMWF_OND_St:Aug / CHIRPS_Obs_OND' )

# Adding grid
plt.grid(True, axis='y')

# Displaying the plot
plt.show()
```
