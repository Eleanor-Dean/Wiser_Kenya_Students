```python
import numpy as np
import xarray as xr
from Data import location, terciles
da = location('Kenya_Hindcast')
obs = location('Kenya_Obs_Chirps')
```

**Change the "below" to "above","below" or "average" to veiw the alternative graphs**

```python
import matplotlib.pyplot as plt

hit_rate,false_alarm_rate = terciles(da,obs,"average") #above,below,average

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(false_alarm_rate, hit_rate, marker='o', color='blue', label='ROC Curve')
plt.plot([0, 1], [0, 1], linestyle='--', color='black', label='Diagonal')

# Add labels, title, legend, and grid
plt.xlabel('False Alarm Rate')
plt.ylabel('Hit Rate')
plt.title('Relative Operating Characteristics (ROC) diagram:\nECMWF_OND_St:Aug / CHIRPS_Obs_OND / (1993-2016)')
plt.legend()
plt.grid(True)  

# Show the plot
plt.show()
```
